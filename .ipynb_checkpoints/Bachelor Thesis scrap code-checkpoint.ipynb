# Fairness metrics

### Statistical Parity Difference

```python
def statistical_parity_difference(y_pred_a, y_pred_b):    
    return y_pred_a.mean() - y_pred_b.mean()
```

### Conditional statistical parity difference

```python
def create_conditional_set(data, attributes):
    for att in attributes:
        data = data.loc[data[att] == attributes[att]]
    return data

def conditional_statistical_parity_difference(X_test, X_test_a, X_test_b, L, model):
    attributes = {}
    for l in L:
        attributes[l] = mode(X_test[l])
        
    X_test_a_cond = create_conditional_set(X_test_a, attributes)
    X_test_b_cond = create_conditional_set(X_test_b, attributes)
    
    y_pred_a_cond = model.predict(X_test_a_cond)
    y_pred_b_cond = model.predict(X_test_b_cond)
    
    cspd = statistical_parity_difference(y_pred_a_cond, y_pred_b_cond)
    return cspd

    
```

### Equal opportunity difference

```python
def equal_opportunity_difference(y_pred_a, y_pred_b, y_test_a, y_test_b): 
    #False negative rate for group a
    [[tp_a, fp_a], [fn_a, tn_a]] = metrics.confusion_matrix(y_test_a, y_pred_a)
    FNR_a = fn_a/(fn_a+tp_a)

    #False negative rate for group b
    [[tp_b, fp_b], [fn_b, tn_b]] = metrics.confusion_matrix(y_test_b, y_pred_b)
    FNR_b = fn_b/(fn_b+tp_b)

    #Equal opportunity difference
    eod = FNR_a-FNR_b
    return eod
```

### Average Odds Difference

```python
def average_odds_difference(y_pred_a, y_pred_b, y_test_a, y_test_b): 
    #False positive rate and true positive rate for group a
    [[tp_a, fp_a], [fn_a, tn_a]] = metrics.confusion_matrix(y_test_a, y_pred_a)
    FPR_a = fp_a/(tn_a+fp_a)
    TPR_a = tp_a/(fn_a+tp_a)

    #False positive rate and true positive rate for male
    [[tp_b, fp_b], [fn_b, tn_b]] = metrics.confusion_matrix(y_test_b, y_pred_b)
    FPR_b = fp_b/(tn_b+fp_b)
    TPR_b = tp_b/(fn_b+tp_b)

    #False positive rate and true positive rate differences
    FPR_diff = FPR_a - FPR_b
    TPR_diff = TPR_a - TPR_b
    
    #Average odds difference
    aod = np.mean((FPR_diff, TPR_diff))
    return aod
```

### Disparate impact

```python
def disparate_impact(y_pred_a, y_pred_b):    
    #Probability of group a to be predicted non recidist
    nonrecid_a = 0
    for i in y_pred_a:
        if i == 0:
            nonrecid_a = nonrecid_a + 1
    P_nonrecid_a = nonrecid_a / y_pred_a.size

    #Probability of group b to be predicted non recidist
    nonrecid_b = 0
    for i in y_pred_b:
        if i == 0:
            nonrecid_b = nonrecid_b + 1
    P_nonrecid_b = nonrecid_b / y_pred_b.size

    #Disparate impact
    di = P_nonrecid_a / P_nonrecid_b
    return di
```

### Theil index

```python
def theil_index(y_pred, y_test):
    #Change y_pred and y_true to be 1 if they have the desired value (being 0)
    y_pred = (y_pred == 0).astype(np.float64)
    y_true = (y_test == 0).astype(np.float64)
    
    #Create array of the benefit
    b = 1 + y_pred - y_true
    
    #Theil index
    ti = np.mean(np.log((b / np.mean(b))**b) / np.mean(b))
    return ti

theil_index(y_pred, y_test)
```

```python
def significance(spd, cspd, eod, aod, di):
    if(spd<-0.1 or spd>0.1):
        print('The statistical parity difference is significant: ', spd )
        
    if(cspd<-0.1 or cspd>0.1):
        print('The conditional statistical parity difference is significant: ', cspd )

    if(eod<-0.1 or eod>0.1):
        print('The equal opportunity difference is significant: ', eod )
    
    if(aod<-0.1 or aod>0.1):
        print('The average odds difference is significant: ', aod )

    if(di<-0.8 or di>1.2):
        print('The disparate impact is significant: ', di )
```

```python
def split_data(X_train, y_train, X_test, y_test, model, att, val1=1, val2=0):
    #Split the data on protected attribute 
    X_train_a = X_train.loc[X_train[att] == val1] 
    y_train_a = y_train.loc[X_train[att] == val1]

    X_train_b = X_train.loc[X_train[att] == val2] 
    y_train_b = y_train.loc[X_train[att] == val2]

    X_test_a = X_test.loc[X_test[att] == val1] 
    y_test_a = y_test.loc[X_test[att] == val1]

    X_test_b = X_test.loc[X_test[att] == val2] 
    y_test_b = y_test.loc[X_test[att] == val2]
    
    if np.shape(X_test_a)[0]==0 or np.shape(X_test_a)[0]==0:
        print(att)
        
    #Predict recidivism
    y_pred_a = model.predict(X_test_a)
    y_pred_b = model.predict(X_test_b)
    
    return X_train_a, y_train_a, X_test_a, y_test_a, y_pred_a, X_train_b, y_train_b, X_test_b, y_test_b, y_pred_b
```

```python
def compute_spd_shap(attribute, val1=1, val2=0):
    #Split data
    X_train_a, y_train_a, X_test_a, y_test_a, y_pred_a, X_train_b, y_train_b, X_test_b, y_test_b, y_pred_b =  split_data(X_train, y_train, X_test, y_test, model, attribute, val1=val1, val2=val2)
    #Compute spd
    spd = statistical_parity_difference(y_pred_a, y_pred_b)
    #Get global SHAP
    att_idx = X_train.columns.get_loc(attribute)
    shap = global_shap_values[att_idx]
    return spd, shap

#Get all columns with binary values
bin_columns = []
nonbin_columns = []
for col in X_train.columns:
    binary = True
    for i in X_train[col]:
        if i !=1 and i!=0:
            binary = False
    if binary == True:
        bin_columns.append(col)
    else:
        nonbin_columns.append(col)

#Make all non binary columns binary
for col in nonbin_columns:
    median_val = median(X_train[col])
    col_binary_x_train = []
    for i in X_train[col]:
        if i <= median_val:
            col_binary_x_train.append(0)
        else:
            col_binary_x_train.append(1)
    col_binary_x_test = []
    for i in X_test[col]:
        if i <= median_val:
            col_binary_x_test.append(0)
        else:
            col_binary_x_test.append(1)
    X_train[col] = col_binary_x_train
    X_test[col] = col_binary_x_test
    bin_columns.append(col)

model.fit(X_train, y_train)
shap_values = shap.TreeExplainer(model).shap_values(X_train)
global_shap_values = shap_values[1].mean(0)

#Compute the SPD and SHAP for all these attributes
spd_vals = np.zeros(np.shape(bin_columns)[0])
shap_vals = np.zeros(np.shape(bin_columns)[0])
for idx, col in enumerate(bin_columns):
    spd_vals[idx], shap_vals[idx] = compute_spd_shap(col)

#Plot the SPD and SHAP
plt.plot(spd_vals, shap_vals, 'ro')
plt.xlabel('spd')
plt.ylabel('shap')
plt.show()

corrcoef = np.corrcoef(spd_vals, shap_vals)
print(corrcoef)
```

```python
def bar_plot(att, val, x_label, size):
    plt.rcdefaults()
    fig, ax = plt.subplots(figsize=size)
    ax.barh(np.arange(len(att)), val, align='center', height=0.8,)
    ax.set_yticks(np.arange(len(att)))
    ax.set_yticklabels(att)
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel(x_label)

    plt.show()
    
```
